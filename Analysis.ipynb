```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # Import NumPy for linear algebra operations
import pandas as pd # Import pandas for data processing and CSV file I/O
```

```python
# Load the training data from the specified CSV file
train_data = pd.read_csv("C:\\phtirado\\Python\\KaggleTitanicChallenge\\Data\\train.csv")
train_data.head() # Display the first few rows of the training data
```

```python
# Load the test data from the specified CSV file
test_data = pd.read_csv("C:\\phtirado\\Python\\KaggleTitanicChallenge\\Data\\test.csv")
test_data.head() # Display the first few rows of the test data
```

```python
# Select all female passengers and their survival status from the training data
women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women) # Calculate the survival rate for women

print("% of women who survived:", rate_women) # Print the survival rate for women
```

```python
# Select all male passengers and their survival status from the training data
men = train_data.loc[train_data.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men) # Calculate the survival rate for men

print("% of men who survived:", rate_men) # Print the survival rate for men
```

```python
from sklearn.ensemble import RandomForestClassifier # Import the RandomForestClassifier

y = train_data["Survived"] # Set the target variable (Survived column)

features = ["Pclass", "Sex", "SibSp", "Parch"] # Define the features to use
X = pd.get_dummies(train_data[features]) # Convert categorical variables to dummy/indicator variables for training data
X_test = pd.get_dummies(test_data[features]) # Convert categorical variables to dummy/indicator variables for test data

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1) # Create a Random Forest model
model.fit(X, y) # Train the model on the training data
predictions = model.predict(X_test) # Predict survival on the test data

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions}) # Create a DataFrame for submission
output.to_csv('submission.csv', index=False) # Save the predictions to a CSV file
print("Your submission was successfully saved!") # Print confirmation message
```
